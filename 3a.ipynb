```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
y = np.arange(-1,1,0.001)
s1 = 1/(1+np.exp(-(500*y+30)))
s2 = 1/(1+np.exp(-(500*y-30)))
s3 = s1 - s2
```

```python
plt.ylabel("s1(y)")
plt.xlabel("y")
plt.plot(y,s1)

plt.show()
```
